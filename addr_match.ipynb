```python
#@title EPC (Environmental Performance Certificates) in chunk by PySpark
# (credit: https://epc.opendatacommunities.org/)
import os
import pandas as pd
from concurrent.futures import ProcessPoolExecutor

def read_certificates(file_path):
    la_name = file_path.split(os.sep)[-2].split('-')[-1]
    chunksize = 10000 #read large CSV files in smaller chunks and process them one at a time. This can help reduce memory usage
    dfs = []

    for chunk in pd.read_csv(file_path, chunksize=chunksize, low_memory=False):
        chunk['la_name'] = la_name
        chunk = optimize_datatypes(chunk)
        dfs.append(chunk)

    return pd.concat(dfs, ignore_index=True)

def optimize_datatypes(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            num_unique_values = len(df[col].unique())
            num_total_values = len(df[col])
            if num_unique_values / num_total_values < 0.5:
                df[col] = df[col].astype('category')
        elif df[col].dtype == 'int64':
            df[col] = pd.to_numeric(df[col], downcast='integer')
        elif df[col].dtype == 'float64':
            df[col] = pd.to_numeric(df[col], downcast='float')
    return df


def find_csv_files(root_dir):
    csv_files = []
    
    for subdir, dirs, files in os.walk(root_dir):
        for file in files:
            if file == 'certificates.csv':
                file_path = os.path.join(subdir, file)
                csv_files.append(file_path)
                
    return csv_files

root_dir = 'epc' 
# root_dir = '/content/addr_data/epc' 
csv_files = find_csv_files(root_dir)

# with ProcessPoolExecutor() as executor: # parallelisation
#     all_dfs = list(executor.map(read_certificates, csv_files))

# epc = pd.concat(all_dfs, ignore_index=True)
# epc.to_parquet('combined_certificates.parquet')

epc = spark.read.csv(csv_files, header = True)
```

```python
#@title pyspark.ml.feature ver | 6min + 6min
def make_tfidf(df):
  from pyspark.ml.feature import HashingTF, IDF, Tokenizer, NGram

  # Tokenization
  tokenizer = Tokenizer(inputCol="addr", outputCol="words")
  df_words = tokenizer.transform(df)

  # N-grams
  ngram = NGram(n=3, inputCol="words", outputCol="ngrams")  # replace 2 with your desired n-gram length
  df_ngrams = ngram.transform(df_words)

  # Term Frequency
  hashingTF = HashingTF(inputCol="ngrams", outputCol="rawFeatures")
  tf = hashingTF.transform(df_ngrams)

  # Inverse Document Frequency
  idf = IDF(inputCol="rawFeatures", outputCol="features")
  idfModel = idf.fit(tf)
  tfidf = idfModel.transform(tf) # Not a matrix yet

  return tfidf

tfidf_epc = make_tfidf(epc)
pyspark.sql.DataFrameWriter(tfidf_epc).saveAsTable('test', format='parquet', mode='overwrite',path='/content/drive/MyDrive/Addr/epcparquet')
```

## Optimized Cosine Similarity  by  sparse_dot_topn 
####  Created by [ING Wholesale Banking Advanced Analytics team](https://medium.com/wbaa/what-does-ing-wb-advanced-analytics-do-707a09175530)

```python
# calculate the similarity between two vectors of TF-IDF values the Cosine Similarity is usually used.
# result matrix in a very sparse terms and Scikit-learn deals with this nicely by returning a sparse CSR matrix.

def awesome_cossim_top(A, B, ntop, lower_bound=0):
    # force A and B as a CSR matrix.
    # If they have already been CSR, there is no overhead
    A = A.tocsr()
    B = B.tocsr()
    M, _ = A.shape
    _, N = B.shape
 
    idx_dtype = np.int32
 
    nnz_max = M*ntop
 
    indptr = np.zeros(M+1, dtype=idx_dtype)
    indices = np.zeros(nnz_max, dtype=idx_dtype)
    data = np.zeros(nnz_max, dtype=A.dtype)

    ct.sparse_dot_topn(
        M, N, np.asarray(A.indptr, dtype=idx_dtype),
        np.asarray(A.indices, dtype=idx_dtype),
        A.data,
        np.asarray(B.indptr, dtype=idx_dtype),
        np.asarray(B.indices, dtype=idx_dtype),
        B.data,
        ntop,
        lower_bound,
        indptr, indices, data)

    return csr_matrix((data,indices,indptr),shape=(M,N))


def PySpark_awesome_cossim_top(A, B, ntop, lower_bound=0):
    M, _ = A.shape
    _, N = B.shape

    idx_dtype = np.int32

    nnz_max = M*ntop
 
    indptr = np.zeros(M+1, dtype=idx_dtype)
    indices = np.zeros(nnz_max, dtype=idx_dtype)
    data = np.zeros(nnz_max, dtype=A.dtype)

    ct.sparse_dot_topn(
        M, N, np.asarray(A.indptr, dtype=idx_dtype),
        np.asarray(A.indices, dtype=idx_dtype),
        A.data,
        np.asarray(B.indptr, dtype=idx_dtype),
        np.asarray(B.indices, dtype=idx_dtype),
        B.data,
        ntop,
        lower_bound,
        indptr, indices, data)
    

    return csr_matrix((data,indices,indptr),shape=(M,N))
```

```python
# unpacks the resulting sparse matrix

def get_matches_df(sparse_matrix, name_vector, top=100):
    non_zeros = sparse_matrix.nonzero()
    
    sparserows = non_zeros[0]
    sparsecols = non_zeros[1]
    
    if top:
        nr_matches = top
    else:
        nr_matches = sparsecols.size
    
    left_side = np.empty([nr_matches], dtype=object)
    right_side = np.empty([nr_matches], dtype=object)
    similairity = np.zeros(nr_matches)
    
    for index in range(0, nr_matches):
        left_side[index] = name_vector[sparserows[index]]
        right_side[index] = name_vector[sparsecols[index]]
        similairity[index] = sparse_matrix.data[index]
    
    return pd.DataFrame({'left_side': left_side,
                          'right_side': right_side,
                           'similairity': similairity})
```

```python
# store the  matches into new dataframe called matched_df and printing 10 samples

matches_df = get_matches_df(matches, room_types, top=200)
matches_df = matches_df[matches_df['similairity'] < 0.99999] # For removing all exact matches
matches_df.sample(10)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
from itertools import product

def get_best_match(string1, string2, cutoff=80):
    """Returns the best fuzzy match score between two strings."""
    return fuzz.token_sort_ratio(string1, string2)

def match_addresses(df, address_cols=['ADDRESS1', 'ADDRESS2', 'ADDRESS3', 'POSTCODE']):
    """Matches addresses in a pandas DataFrame using fuzzy matching.

    Args:
        df (pandas.DataFrame): DataFrame containing address data.
        address_cols (list): List of column names containing address data.

    Returns:
        pandas.DataFrame: DataFrame with additional 'Match' column indicating whether an address has a fuzzy match.
    """
    # create a new column to store fuzzy match scores
    df['Match'] = False
    
    # create a cartesian product of the rows to compare all possible pairs
    row_pairs = list(product(df.index, repeat=2))
    
    for idx1, idx2 in row_pairs:
        # skip matching the same row
        if idx1 == idx2:
            continue
        # calculate fuzzy match score between the two rows
        score = sum(get_best_match(df.loc[idx1, col], df.loc[idx2, col]) for col in address_cols) / len(address_cols)
        # set 'Match' to True if score is above cutoff
        if score >= 80:
            df.loc[idx1, 'Match'] = True
            df.loc[idx2, 'Match'] = True
            
    return df
df = match_addresses(chi, address_cols=['ADDRESS1', 'ADDRESS2', 'ADDRESS3', 'POSTCODE'])
```
